```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sm=pd.read_csv('titanic.csv')
sm
```

```python
temp=sm.head(30)
temp
```

```python
plt.figure(figsize=(10,5))
plt.plot(temp['embark_town'],temp['fare'],marker='o',)
plt.title("Embark Town VS Age")
plt.xlabel('embark_town')
plt.ylabel('fare')
plt.grid(True)
plt.show()
```

```python
data = temp['survived']
label = temp['sex']
plt.pie(data,labels=label)
plt.show()
```
